This notebook explores the daily dialogue dataset and shows how it is used.

```python
#First we fix the relative imports
import os
import sys
# M
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
from daily_dialog.UtterancesDataset import UtterancesDataset
from utils.utils import decode, batch_decode
```

```python
# Now we wil load a tokenizer as the dataset expects a dataset because it sorts based on the actual tokens
from daily_dialog.DialogTokenizer import get_daily_dialog_tokenizer
my_tokenizer = get_daily_dialog_tokenizer(tokenizer_location='../daily_dialog/tokenizer.json', )
```

```python
size=100 # We take a subset of the dataset, just for convienence
dataset_train = UtterancesDataset(my_tokenizer, subsets="start", split="train", size=size)
```

```python
### NExt we take the first item from the dataset, which contains a tuple, uterrance, respopnse with the ids of the tokens
context_ids, response_ids = dataset_train[1]
```

```python
# Use the tokenizer to see what the actual utterance is. We use the custom decode function to make the decoded text pretty. 
context = decode(my_tokenizer, context_ids)
print(context)
response = decode(my_tokenizer, response_ids)
print(response)
```

```python
## Next we look at what happens when we put it in a dataloader, we use the collate function for this.
from torch.utils.data import DataLoader
batch_size=32
#Note that we do not want to shuffle as this is already done inside the dataset itself (similar lenghts )
# You can set shuffle=True to see the difference
dataloader_train = DataLoader(dataset_train, batch_size=batch_size,  collate_fn=UtterancesDataset.get_collate_fn())
```

```python
for batch in dataloader_train:
    print(batch[0].shape)
```

```python
#Next we want to batch decode to see how the sentences look
import torch
for batch in dataloader_train:
    context_response = torch.cat([batch[0], batch[1]], dim=-1)
    print(batch_decode(my_tokenizer, context_response))
```
